Body segment parameters (BSP), such as length, area, volume, mass, density, center of mass, moment of inertia, and center of volume, are fu inndamental for the study of the biomechanics of human movement.   
In most cases, one can not measure mass, center of mass and rotational inertia of each segment of a subject and then these quantities have to be estimated by an indirect method based on data of a group of subjects.  

The main indirect methods are based in data of: cadavers (e.g. Dempster model), body image scanning of living subjects (e.g., Zatsiorsky model), and geometric model (e.g., Hanavan model). See [http://kwon3d.com/theory/bsp.html](http://kwon3d.com/theory/bsp.html) and [Human Body Properties Database: Body Segment Parameters](http://riodb.ibase.aist.go.jp/dhbodydb/properties/m/index-e.html) for more information on these models and others.

Some key concepts in this area are:   

+ **Length, area, volume**: Dimensional measures of one-, two-, and three-dimensional geometric objects. SI unit: $m,\;m^2,\;m^3\:$ (scalar).
+ **Mass**: More specifically, inertial mass is a quantitative measure of an object's resistance to acceleration. SI unit: $kg\:$ (scalar).
+ **Density**: density of an object is its mass per unit volume. SI unit: $kg/m^3\:$ (scalar).
+ **Center of mass**: The center of mass is the unique point at the center of a distribution of mass in space that has the property that the weighted position vectors relative to this point sum to zero. SI unit: $m\:$ (vector).
+ **Center of gravity**: Center of gravity is the point in an object around which the resultant torque due to gravity forces vanishes. Near the surface of the earth, where the gravity acts downward as a parallel force field, the center of gravity and the center of mass are the same. SI unit: $m\:$ (vector).
+ **Rotational inertia (or moment of inertia)**: Analogous to mass, it a quantitative measure of the resistance to rotational acceleration about an axis of a distribution of mass in space. $kg\:m^2\:$ (scalar).
 - **Radius of gyration**: It is the distance from the axis that all mass can be concentrated to obtain the same moment of inertia of the object. SI unit: $m\:$ (scalar).
 - **Parallel axis theorem**: According to this theorem, the moment of inertia about a parallel axis is the moment of inertia around the center of mass plus the moment of inertia of the body treated as a point mass at this new axis location.
+ **Center of volume**: Analogous to the center of mass, is the unique point at the center of a distribution of volume in space that has the property that the weighted position vectors relative to this point sum to zero. This concept is important for the understanding of movement in water. SI unit: $m\:$ (vector).
+ **Center of buoyancy**: It is the center of the volume of water which the submerged part of an object displaces. Center of buoyancy is to center of volume as center of gravity is to center of mass. SI unit: $m\:$ (vector).   
+ **Anthropometry**: the study of human body measurements (such as these mentioned above).

In this text, we will see how to estimate some of these parameters using the Dempster (1973) and Zatsiorsky and Seluyanov (1990) models.

For a table with BSP values, typically:   

+ The mass of each segment is given as fraction of the total body mass.   
+ The center of mass (cm) position in the sagital plane of each segment is given as fraction of the segment length with respect to the proximal (p) or distal (d) joint position.
+ The radius of gyration (rg) in the sagital plane of each segment is given as fraction of the segment length with respect to the center of mass (cm) or proximal (p) or distal (d) joint position.

##Center of mass
The mathematical definition for the center of mass (or barycenter) of a system with N objects (a system of particles), each with mass $m_i$ and position $r_i$ is:

$$ \mathbf{r_{cm}} = \frac{1}{M}\sum_{i=1}^N m_{i}\mathbf{r_i} \;\;\;\;\;\;\text{where}\;\;\;\;\;\; M = \sum_{i=1}^N m_{i} $$   

Using the data of the body segment parameters table, the position of the center of mass of a single segment $i$ is (see figure below):

$$r_{i} = r_{i,p} + \text{bsp[i,cmp]} * (r_{i,d}-r_{i,p})$$

Where $r_{i,p}\:$ and $\:r_{i,d}\:$ are the positions of the proximal and distal landmarks used to define the $i$ segment.   
Note that $r$ is a vector and may have more than one dimension. The equation for the center of mass position is valid in each direction and the calculations are performed independently in each direction. In addition, there is no need to include the mass of the segment in the equation above; the mass of the segment is used only when there is more than one segment. 

For example, given the following coordinates (x,y) for the MT2, ankle, knee and hip joints:

```python
r = np.array([[101.1, 1.3], [84.9, 11.0], [86.4, 54.9], [72.1, 92.8]])/100
display(np.around(r, 3))
```

The position of the center of mass of the lower limb is:

```python
M = bspD.ix[['foot','leg','thigh'],'mass'].sum()
rcm_foot  = r[1] + bspD.ix['foot', 'cmp']*(r[0]-r[1])
rcm_leg   = r[2] + bspD.ix['leg',  'cmp']*(r[1]-r[2])
rcm_thigh = r[3] + bspD.ix['thigh','cmp']*(r[2]-r[3])
rcm = (bspD.ix['foot','mass']*rcm_foot + bspD.ix['leg','mass']*rcm_leg + \
       bspD.ix['thigh','mass']*rcm_thigh)/M
display(np.around(rcm, 3))
```

And here is a geometric representation of part of these calculations:

```python
plt.rc('axes',  labelsize=14, linewidth=1.5) 
plt.rc('xtick', labelsize=14) 
plt.rc('ytick', labelsize=14)
plt.rc('lines', markersize=8) 
hfig, hax = plt.subplots(1, 1, figsize=(10,6))

# bones and joints
plt.plot(r[:,0], r[:,1],'b-')
plt.plot(r[:,0],r[:,1],'ko', label='joint')
# center of mass of each segment
plt.plot(rcm_foot[0], rcm_foot[1],'go', label='segment center of mass')
plt.plot(rcm_leg[0], rcm_leg[1],'go', rcm_thigh[0], rcm_thigh[1],'go')
# total center of mass
plt.plot(rcm[0], rcm[1],'ro', label='total center of mass')
hax.legend(frameon=False, loc='upper left', fontsize=14,numpoints=1)
plt.arrow(0, 0, r[3,0], r[3,1],'b', head_width=0.02, overhang=.5, fc="k",ec="k", lw=2,
          length_includes_head=True)
plt.arrow(r[3,0], r[3,1], rcm_thigh[0] - r [3,0], rcm_thigh[1] - r[3,1], head_width=0.02,
      overhang=.5, fc="b", ec="b", lw=2, length_includes_head=True)
plt.arrow(0, 0, rcm_thigh[0], rcm_thigh[1], head_width=0.02, overhang=.5, fc="g", ec="g",
          lw=2, length_includes_head=True)
plt.text(0.30, .57, '$\mathbf{r}_{thigh,p}$', rotation=40, fontsize=20)
plt.text(0.77, .85, '$bsp_{thigh,cmp}*(\mathbf{r}_{i,d}-\mathbf{r}_{i,p})$', fontsize=20, color='b')
plt.text(0.20, .55,
         '$\mathbf{r}_{thigh,cm}=\mathbf{r}_{i,p}+bsp_{i,cmp}*(\mathbf{r}_{i,d}-\mathbf{r}_{i,p})$',
         rotation=30, fontsize=20, color='g')
hax.set_xlim(0,1.1)
hax.set_ylim(0,1.05)
hax.set_xlabel('x [m]')
hax.set_ylabel('y [m]')
hax.set_title('Determination of center of mass', fontsize=20)
hax.grid()
```

### Center of mass of a body with continuous distribution of mass
If the mass distribution of the body is continous over a volume and each element of mass has density $\rho(r),\:$ the position of the center of mass is given by:

$$ \mathbf{r_{cm}} \;=\; \frac{1}{M}\int_0^M \mathbf{r}\:dm \;=\; \frac{1}{M}\int_0^V \mathbf{r}\rho(r)\:dV $$

For example, the center of mass of a uniform rod with total mass M and length L along to its main axis x is:

$$ \mathbf{r_{cm}} \;=\; \frac{1}{M}\int_0^L x\frac{M}{L}\:dx = \frac{1}{M}\frac{M}{L}\left.\frac{x^2}{2}\right|_0^L = \frac{L}{2} $$

For a rod with M = 1 kg and L = 1 m, $r_{cm}$ = 0.5 m.

##Rotational inertia

The mathematical definition for the rotational inertia of a system with N particles rotating around an axis, each with mass $m_i$ and distance $r_i$ to this axis is:

$$ I = \sum_{i=1}^N m_{i}r_{i}^2 $$    

###Radius of gyration

The mathematical definition for the radius of gyration, $r_{rg}$, is:

$$ I = \sum_{i=1}^N m_{i}r_{i}^2 = \left(\sum_{i=1}^N m_{i}\right)*r_{rg}^2 \;\; \Rightarrow \;\; r_{rg} = \sqrt{\frac{I}{\sum_{i=1}^N m_{i}}} = \sqrt{\frac{\sum_{i=1}^N m_{i}r_{i}^2}{\sum_{i=1}^N m_{i}}}$$

The radius of gyration (as a fraction of the segment length) is the quantity that is given in the table of body segment parameters. Because of that, we don't need to sum each element of mass of the segment to calculate its rotational inertia; we just need to take the mass of the segment times the radius or gyration squared.

Using the body segment parameters, the rotational inertia of a single segment $i$ rotating around its own center of mass is (see figure below):

$$I_{i,cm} = M * \text{bsp[i,mass]} * \left(\text{bsp[i,rgcm]} * ||r_{i,d}-r_{i,p}||\right)\:^2 $$

Where $M$ is the total body mass of the subject and $ \:||r_{i,d}-r_{i,p}||\;\; $ is the length of the segment $i$. 

###Parallel axis theorem

The value of the rotational inertia is by definition dependent of the distance of the mass to the axis of rotation. This is the reason the radius of gyration in the table above is given for three different axis (center of mass and proximal and distal joint positions). In case we want the rotational inertia of an object around a different parallel axis than the axis passing through the center of mass, for example of the foot around the hip joint center, we can use the parallel axis theorem:

$$ I_{new\:axis}\;=\;I_{cm}\;+\;md_{new\:axis\:to\:cm}^2 $$

For example, the rotational inertia of the lower limb rotating around the hip joint center considering the coordinates (x,y) for the MT2, ankle, knee and hip joints given above is (suing the parallel axis theorem):

```python
norm = np.linalg.norm
M = 80 # body mass
I = M*(bspD.ix['foot', 'mass']*((bspD.ix['foot', 'rgcm']*norm(r[0]-r[1]))**2+norm(r[3]-rcm_foot )**2)+\
       bspD.ix['leg',  'mass']*((bspD.ix['leg',  'rgcm']*norm(r[1]-r[2]))**2+norm(r[3]-rcm_leg  )**2)+\
       bspD.ix['thigh','mass']*((bspD.ix['thigh','rgcm']*norm(r[2]-r[3]))**2+norm(r[3]-rcm_thigh)**2))

display(Latex(str(np.around(I,3)) + ' kgm$^2$'))
```

### Rotational inertia of a body with continuous distribution of mass

If the mass distribution of the body is continous over a volume and each element of mass has density $\rho(r)\:$, the value of the rotational inertia of the body rotating around an axis passing through the origin is given by:

$$ I = \int_0^M r^2\:dm \;=\; \int_0^V \mathbf{r}^2\rho(r)\:dV $$

For example, the rotational inertia of a uniform rod with total mass M and length L rotating around an axis passing through its center of mass and perpendicular to its main axis x is:

$$ I = \int_{-L/2}^{L/2} x^2\frac{M}{L}\:dx = \frac{M}{L}\left.\frac{x^3}{3}\right|_{-L/2}^{L/2} = M\frac{L^2}{12} $$

For a rod with M = 1 kg and L = 1 m, I = 0.083 kgm$^2$.

If now the axis of rotation passes through one of its extermities and it is parallel to the former axis passing through the center of mass, the rod rotational inertia is:

$$ I = \int_{0}^{L} x^2\frac{M}{L}\:dx = \frac{M}{L}\left.\frac{x^3}{3}\right|_0^L = M\frac{L^2}{3} $$

For a rod with M = 1 kg and L = 1 m, I = 0.333 kgm$^2$.

We could have calculated this rotational inertia using the parallel axis theorem:

$$ I = I_{cm} + md^2_{new\:axis\:to\:cm}\;= M\frac{L^2}{12} + M\left(\frac{L}{2}\right)^2 = M\frac{L^2}{3} $$

### Rotational inertia tensor

The rotational inertia is dependent on the axis around which the body rotates and the axes (system of coordinates) we chose to describe the body (volume) in space. In a general form, considering these two sets of axes in 3D space, the rotational inertia of a body is given by the following tensor:

$$ I = \begin{bmatrix}
I_{xx} & I_{xy} & I_{xz} \\\ 
I_{yx} & I_{yy} & I_{yz} \\\ 
I_{zx} & I_{zx} & I_{zz} 
\end{bmatrix} $$

This matrix is symmetric: $I_{xy}=I_{yx}\:; I_{xz}=I_{zx}\:; I_{yz}=I_{zy}\:.$   

If we choose to calculate the rotational inertia around the three axes of rotational symmetry of the body (called principal axes), the off-diagonal terms of the rotational inertia tensor are zero. When calculating the rotational inertia for the human body segments in motion, it is often preferred to perform the calculations around these principal axes for each segment (local coordinate system) and use only the main diagonal terms of the rotational inertia tensor than perform all the calculations in a global coordinate system and have to use the full rotational inertia tensor.
