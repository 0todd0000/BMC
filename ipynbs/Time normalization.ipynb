Time normalization is usually employed for the temporal alignment of cycle data obtained from different trials with different duration (number of points). The most simple and common method used in Biomechanics and Motor Control is a procedure knwown as the normalization to percent cycle (although it might not be the most adequate in certain cases ([Helwig et al., 2011](http://www.sciencedirect.com/science/article/pii/S0021929010005038))).

In the percent cycle, a fixed number (typically a temporal base from 0 to 100%) of new equaly spaced data is created based on the old data with a mathematical procedure knwon as interpolation.   
**Interpolation** is the estimation of new data points within the range of known data points. This is different from **extrapolation**, the estimation of data points outside the range of known data points.   
Time normalization of data using interpolation is a simple procedure and it doesn't matter if the original data have more or less data points than desired.

For example, consider the following data:

```python
y = [5,  4, 10,  8,  1, 10,  2,  7,  1,  3]
```

The time normalization of these data to represent a cycle from 0 to 100%, with a step of 1% (101 data points) is:

```python
# Import the necessary libraries
from __future__ import division, print_function
import sys, os
sys.path.insert(1, os.path.join(os.path.abspath('..'), 'functions'))  # add as first in pythonpat
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import HTML #, display, Image
```

```python
t  = np.linspace(0, 100, len(y))  # time vector for the original data
tn = np.linspace(0, 100, 101)     # new time vector for the new time-normalized data
yn = np.interp(tn, t, y)          # new time-normalized data
```

The key is the Numpy `interp` function, from its help:   

>interp(x, xp, fp, left=None, right=None)       
>One-dimensional linear interpolation.   
>Returns the one-dimensional piecewise linear interpolant to a function with given values at discrete data-points.

A plot of the data will show what we have done:

```python
plt.figure(figsize=(10,5))
plt.plot(t, y, 'bo-', lw=2, label='original data')
plt.plot(tn, yn, '.-', color=[1, 0, 0, .5], lw=2, label='time normalized')
plt.legend(loc='best', framealpha=.5)
plt.xlabel('Cycle [%]')
plt.show()
```

Nothing wrong with the lines of the code above, but we may want a different interpolation than the linear one used, we may have to deal with missing points in the data that we should be able to interpolate through them (if these missing points are not at the extremities of the data because the interpolation function can not extrapolate data), and we typically have data with multiple columns representing different variables that we want normalize in time all of them.   

So, let's write a more robust and versatile function for the typical use in Biomechanics and Motor Control.   
Here is its help:

```python
from tnorm import tnorm
help(tnorm)
```

Let's test this code:

```python
    >>> y = [1,  9,  8,  5,  4,  5,  8, 10,  2,  7, 10,  4,  7,  8,  6,  2,  5, 6,  2,  2]
    >>> # Default options: cubic spline interpolation passing through
    >>> # each datum, 101 points, and no plot
    >>> yn, tn = tnorm(y)
    print(yn)
    print(tn)
```

```python
    >>> # Plot with other default options
    >>> yn, tn = tnorm(y, show=True)
```

```python
    >>> # Linear interpolation passing through each datum
    >>> yn, tn = tnorm(y, degree=1, smooth=0, mask=None, show=True)
```

```python
    >>> # Cubic spline interpolation with smoothing
    >>> yn, tn = tnorm(y, degree=3, smooth=1, mask=None, show=True)
```

```python
    >>> import numpy as np
    >>> x = np.linspace(-3, 3, 100)
    >>> y = np.exp(-x**2) + np.random.randn(100)/10
    >>> # Cubic spline interpolation with smoothing and 50 points
    >>> yn, tn = tnorm(y, step=-50, degree=3, smooth=1, show=True)
```

```python
    >>> # Deal with missing data (use NaN as mask)
    >>> y[0] = np.NaN # first point is also missing
    >>> y[30: 41] = np.NaN # make other 10 missing points
    >>> yn, tn = tnorm(y, step=-50, degree=3, smooth=1, show=True)
```

So, we apply the time normalization for all trials we have (of the same subject or different subjects) and then we are ready to calculate the mean and standard deviation curves across trials to express the average pattern and dispersion of the movement.
