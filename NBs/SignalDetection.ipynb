An important operation in signal processing is to detect or identify the occurrence of a particular event in the data, for example, its beginning and ending, which can also ne referred as the onset period. For instance, in the analysis of electromyographic (EMG) data, one of the first tasks is to detect the onset period of muscle activation based on the amplitude of the EMG signal which typically has noise. Another example would be to detect the beginning and ending of a discrete movement such as reaching something with the hand. This detection can be done manually by showing the amplitude of the EMG data or the hand velocity on a graph and the user would manually select the events of interest, but a lot of research has been conducted to propose automatic methods to perform this task. Signal detection is an active research area in signal processing and it is relevant to all types of signals.

## Amplitude-threshold method

One of the simplest methods to automatically detect the onset of a signal is based on amplitude threshold, where the signal is considered to be 'on' when it is above a certain threshold. This threshold can be proportional to the amplitude of the baseline (the part of the data that we know there is no real signal, only noise), for example, a threshold equals to two or three times the standard deviation of the baseline (a common simple procedured employed in the analysis of EMG data). Other way to set the threshold would be as a percentage value of the maximum or peak of the data, for example, we would define the onset period as the signal above 5% of the peak hand velocity. To avoid the detection of spikes or transients in the data, a second threshold can be used imposing that the signal should be above the threshold for at least a certain period (number of samples).  

Here is a function called `onset_detection.py` (code at the end of this text) that performs such onset detection based on the amplitude-threshold method with a parameter to specify a minimum number of samples above threshold to detect as onset and another parameter to specify the minimum number of samples (continuous or not) below threshold that will be ignored in the detection of data greater or equal to threshold.

## Function `onset_detection.py`

```python
#!/usr/bin/env python

"""Detects onset in data based on amplitude threshold."""

from __future__ import division, print_function
import numpy as np

__author__ = 'Marcos Duarte, https://github.com/duartexyz/BMC'
__version__ = 'onset_detection.py v.1 2014/06/05'


def onset_detection(data, threshold, n_above=1, n_below=0, show=False, ax=None):
    """Detects onset in data based on amplitude threshold.

    Parameters
    ----------
    x         : 1D array_like
                data.
    threshold : number
                minimum amplitude of the data to detect.
    n_above   : number, optional (default = 1)
                minimum number of continuous samples greater than or equal to
                threshold to detect (but see the parameter n_below).
    n_below   : number, optional (default = 0)
                minimum number of samples (continuous or not) below threshold
                that will be ignored in the detection of data >= threshold.
    show      : bool, optional (default = True)
                True (1) plots data in matplotlib figure, False (0) don't plot.
    ax        : a matplotlib.axes.Axes instance, optional (default = None).

    Returns
    -------
    inds      : 1D array_like [indi, indf]
                initial and final indeces of the onset events.

    Notes
    -----
    You might have to tune the parameters according to the signal-to-noise
    characteristic of the data.

    See this IPython Notebook [1]_.

    References
    ----------
    .. [1] http://nbviewer.ipython.org/github/duartexyz/BMC/blob/master/OnsetDetection.ipynb

    Examples
    --------
    >>> import numpy as np
    >>> from onset_detection import onset_detection
    >>> x = np.random.randn(100)
    >>> onset_detection(x, threshold=0, n_above=10, n_below=1, show=True)
    >>> x = np.random.randn(200)/10 
    >>> x[51:151] += np.hstack((np.linspace(0,1,50), np.linspace(1,0,50)))
    >>> inds = onset_detection(x, np.std(x[:50]), 10, 0, True)
    >>> inds
    """

    data = np.atleast_1d(data)    
    # deal with NaN's (by definition, NaN's are not greater than threshold)
    data[np.isnan(data)] = -np.inf    
    # indices of data greater than or equal to threshold
    inds = np.nonzero(data >= threshold)[0]
    if inds.size:
        # initial and final indexes of continuous data
        inds = np.vstack((inds[np.diff(np.hstack((-np.inf, inds))) > n_below+1], \
                          inds[np.diff(np.hstack((inds, np.inf))) > n_below+1])).T
        # indexes of continuous data longer than or equal to n_above
        inds = inds[inds[:, 1]-inds[:, 0] >= n_above, :]

    if not inds.size:
        inds = np.array([])  # standardize inds shape
    if show and data.size > 1:  # don't waste my time ploting one datum
        _plot(data, threshold, n_above, n_below, inds, ax)

    return inds


def _plot(data, threshold, n_above, n_below, inds, ax):
    """Plot results of the cogve function, see its help."""
    try:
        import matplotlib.pyplot as plt
    except ImportError:
        print('matplotlib is not available.')
    else:
        if ax is None:
            fig, ax = plt.subplots(1, 1, figsize=(8, 4))

        if inds.size:
            for (indi, indf) in inds:
                ax.plot(range(indi, indf+1), data[indi:indf+1], 'r', lw=1)
                ax.axvline(x=indi, color='b', lw=1, ls='--')
                ax.axvline(x=indf, color='b', lw=1, ls='--')

            inds = np.vstack((np.hstack((0, inds[:, 1])),
                              np.hstack((inds[:, 0], data.size-1)))).T

            for (indi, indf) in inds:
                ax.plot(range(indi+1, indf), data[indi+1:indf], 'k', lw=1)

        else:
            ax.plot(data, 'k', lw=1)
            ax.axhline(y=threshold, color='r.', lw=1, ls='-')

        ax.set_xlim(-.02*data.size, data.size*1.02)
        ax.set_xlabel('Data #', fontsize=14)
        ax.set_ylabel('Amplitude', fontsize=14)
        ax.set_title('Onset detection (threshold=%.3g, n_above=%d, n_below=%d)'\
                     %(threshold, n_above, n_below))
        #plt.grid()
        plt.show()
```
